```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
try:
    file_path = r'C:\Users\simon\Downloads\squirrel_diet.csv'
    squirrel_data = pd.read_csv(file_path)
except FileNotFoundError:
    squirrel_data = pd.read_csv('/datasets/squirrel_diet.csv')

# Map diet numbers to labels for better readability in the plot
squirrel_data['diet_label'] = squirrel_data['diet'].map({1: 'A', 2: 'B', 3: 'C'})

# Map sex numbers to labels for plot legend
squirrel_data['sex_label'] = squirrel_data['sex'].map({0: 'Male', 1: 'Female'})

# Add jitter to the x-axis to space out the points
squirrel_data['diet_jitter'] = squirrel_data['diet'] + np.random.normal(0, 0.05, size=len(squirrel_data))

# Create the scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=squirrel_data, x='diet_jitter', y='weight', hue='sex_label',
                palette={'Female': 'magenta', 'Male': 'green'}, s=100)

# Customize labels and title
plt.title("Red Squirrel Diet Study")
plt.xlabel("Diet Label")
plt.ylabel("Adult Weight / lbs")
plt.xticks([1, 2, 3], ['A', 'B', 'C'])  # Ensure the correct labels are shown

# Display the plot
plt.show()
```

# Title: Red Squirrel Diet Study Analysis
# Description: This project visualizes the relationship between different diet types and the weight of adult red squirrels. 
# The data also includes the sex of the squirrels, allowing us to observe any differences between males and females.

Observations
Distribution of Diets: The plot shows data points distributed across three diet categories (A, B, and C). The jitter on the x-axis helps visualize the spread of data points for each diet.

Weight Variations: The weight of adult squirrels appears to vary across different diet categories. Diet A and Diet C show a wider range of weights, whereas Diet B has a more concentrated cluster of weights.

Sex Differences: There does not seem to be a clear difference in weight distribution between male and female squirrels within the same diet category.
Outliers: There are a few data points that appear to be outliers with weights significantly higher or lower than the rest, especially in Diet A and Diet C.

Sample Size: The number of data points in each diet category varies, with Diet A and Diet C having more samples compared to Diet B.
